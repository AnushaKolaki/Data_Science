# Marker edge color

```python
a=np.array([1,2,3,4,5])
b=np.array([10,25,35,40,50])
plt.plot(a,b,marker='^',ms=20,mec='r',color='y')
plt.show()
```

# Grid

```python
a=np.array([1,2,3,4,5])
b=np.array([10,25,35,40,50])
plt.plot(a,b,marker='^',ms=20,mec='r',color='y')
plt.grid()
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([10,25,35,40,50])
plt.plot(a,b,marker='^',ms=20,mec='r',color='y')
plt.grid(color='r')
plt.show()
```

# ls Line Style

```python
a=np.array([1,2,3,4,5])
b=np.array([10,25,35,40,50])
plt.plot(a,b,marker='^',ms=20,mec='r',color='y',ls='--')
plt.grid(color='r')
plt.show()
```

# Grid Line Style

```python
a=np.array([1,2,3,4,5])
b=np.array([10,25,35,40,50])
plt.plot(a,b,marker='^',ms=20,mec='r',color='y',ls='--')
plt.grid(color='r',ls='--')
plt.show()
```

# Line width

```python
a=np.array([1,2,3,4,5])
b=np.array([10,25,35,40,50])
plt.plot(a,b,marker='^',ms=20,mec='r',color='y',ls='--',linewidth=5)
plt.grid(color='r',ls='--')
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([10,25,35,40,50])
plt.plot(a,b,marker='^',ms=20,mec='r',color='y',ls='--',linewidth=5)
plt.grid(color='r',ls='--',linewidth=2)
plt.show()
```
